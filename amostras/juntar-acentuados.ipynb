```python
import unicodedata

def shave_marks(txt):
    """Remove all diacritic marks"""
    norm_txt = unicodedata.normalize('NFD', txt)
    shaved = ''.join(c for c in norm_txt
                     if not unicodedata.combining(c))
    return unicodedata.normalize('NFC', shaved)

pares = []
for nome_ac in set(nomes):
    nome = shave_marks(nome_ac)
    if nome_ac != nome:
        pares.append((nome, nome_ac))

with open('nomes-acentuados.tsv', 'w', encoding='utf8') as fp:
    for nome, nome_ac in sorted(pares):
        line = f'{nome}\t{nome_ac}'
        print(line)
        fp.write(line + '\n')
```
